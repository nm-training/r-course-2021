```python
import pandas as pd
import numpy as np
q6a = input("Do you want to import sample names? Press 'y' or 'n':")
q6a = q6a.lower()
def TH_for_50(F,I):
    Total = (F-I)*20
    return Total
def Ca_for_50(Fi,In):
    Tot = (Fi-In)*8
    return Tot
def TH_for_25(final, initial):
    tot = (final-initial)*20*2
    return tot
def Ca_for_25(fii, inn):
    ttt = (fii-inn)*8*2
    return ttt
def list_append(empty_list, loop_value):
    val =0
    for i in range(loop_value):
        empty_list.append(val+1)
        val= val+1
    return empty_list
def all_calculation_50(q6f):
        daka1 ={}
        daka2 ={}
        daka3 ={}
        daka4 ={}
        daka4 ={}
        daka5 ={}
        daka6 ={}
        daka7 ={}
        for a_v in range(len(q6f)):
            print(" ")
            print(" ")
            print(("Enter values for Total Hardness only.").upper())
            print("For", q6f[a_v],":")
            q6g = float(input("Initial value:"))
            q6h = float(input("Final value:"))
            q6Hard = TH_for_50(q6h,q6g)
            daka1[q6f[a_v]] = q6f[a_v]
            daka2[q6f[a_v]] = q6g
            daka3[q6f[a_v]] = q6h
            daka4[q6f[a_v]] = q6Hard
        sE1 = pd.Series(daka1)
        sE2 = pd.Series(daka2)
        sE3 = pd.Series(daka3)
        sE4 = pd.Series(daka4)
        q6i = input("Do you want to calculate the calcium content as well?: Press'y' or 'n':")
        q6i =  q6i.lower()
        if q6i == 'y':
            for a_u in range(len(q6f)):
                print(" ")
                print(" ")
                print(("Enter values for Calcium only.").upper())
                print("For", q6f[a_u],":")
                q6j = float(input("Initial value:"))
                q6k = float(input("Final value:"))
                q6Ca = Ca_for_50(q6k,q6j)
                daka6[q6f[a_u]] = q6j
                daka7[q6f[a_u]] = q6k
                daka5[q6f[a_u]] = q6Ca
            sE5 = pd.Series(daka6)
            sE6 = pd.Series(daka7)
            sE7 = pd.Series(daka5)
            dfTH = pd.DataFrame({"Sample:":daka1, "F_TH:":daka3, "I_TH:":daka2, "Total Hardness:":daka4, "F_ca:":daka7, "I_ca:":q6j, "Calcium:":daka5})
            q6l = input("Do you want to calculate the magnesium content as well? Press 'y' or 'n':")
            q6l =q6l.lower()
            if q6l == 'y':
                dfTH['Magnesium'] = dfTH["Total Hardness:"] - dfTH["Calcium:"]
                return dfTH
            elif q6l == 'n':
                dfTH =dfTH
                return dfTH
        elif q6i == 'n':
            dfTH = pd.DataFrame({"Sample:":daka1, "F_TH:":daka3, "I_TH:":daka2, "Total Hardness:":daka4})
            return dfTH
        

def all_calculation_25(q6f):
        daka1 ={}
        daka2 ={}
        daka3 ={}
        daka4 ={}
        daka4 ={}
        daka5 ={}
        daka6 ={}
        daka7 ={}
        for a_v in range(len(q6f)):
            print(" ")
            print(" ")
            print(("Enter values for Total Hardness only.").upper())
            print("For", q6f[a_v],":")
            q6g = float(input("Initial value:"))
            q6h = float(input("Final value:"))
            q6Hard = TH_for_25(q6h,q6g)
            daka1[q6f[a_v]] = q6f[a_v]
            daka2[q6f[a_v]] = q6g
            daka3[q6f[a_v]] = q6h
            daka4[q6f[a_v]] = q6Hard
        sE1 = pd.Series(daka1)
        sE2 = pd.Series(daka2)
        sE3 = pd.Series(daka3)
        sE4 = pd.Series(daka4)
        q6i = input("Do you want to calculate the calcium content as well?: Press'y' or 'n':")
        q6i =  q6i.lower()
        if q6i == 'y':
            for a_u in range(len(q6f)):
                print(" ")
                print(" ")
                print(("Enter values for Calcium only.").upper())
                print("For", q6f[a_u],":")
                q6j = float(input("Initial value:"))
                q6k = float(input("Final value:"))
                q6Ca = Ca_for_25(q6k,q6j)
                daka6[q6f[a_u]] = q6j
                daka7[q6f[a_u]] = q6k
                daka5[q6f[a_u]] = q6Ca
            sE5 = pd.Series(daka6)
            sE6 = pd.Series(daka7)
            sE7 = pd.Series(daka5)
            dfTH = pd.DataFrame({"Sample:":daka1, "F_TH:":daka3, "I_TH:":daka2, "Total Hardness:":daka4, "F_ca:":daka7, "I_ca:":q6j, "Calcium:":daka5})
            q6l = input("Do you want to calculate the magnesium content as well? Press 'y' or 'n':")
            q6l =q6l.lower()
            if q6l == 'y':
                dfTH['Magnesium'] = dfTH["Total Hardness:"] - dfTH["Calcium:"]
                return dfTH
            elif q6l == 'n':
                dfTH =dfTH
                return dfTH
        elif q6i == 'n':
            dfTH = pd.DataFrame({"Sample:":daka1, "F_TH:":daka3, "I_TH:":daka2, "Total Hardness:":daka4})
            return dfTH
        
    
    
    
    
if q6a == 'y':
    q6b = input("Enter the name of file from which you want to import: ")
    q6c= int(input("Enter the volume for these samples: "))
    q6d = pd.read_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(q6b))
    q6e = list(q6d.columns)
    q6ff = list(q6d[q6e[1]])
    if q6c == 50:
        df_all = all_calculation_50(q6ff)
    elif q6c ==25:
        df_all = all_calculation_25(q6ff)
if q6a == 'n':
    q7a = int(input("Enter the total number of samples:"))
    q7b = int(input("Enter the volume of samples: "))
    q6ff = []
    for i in range(q7a):
        q7c = (input("Enter the name of the sample:")).upper()
        q6ff.append(q7c)
    if q7b == 50:
        df_all = all_calculation_50(q6ff)
    elif q7b == 25:
        df_all = all_calculation_25(q6ff)
print(df_all)
q7d = input("Save the file with initial and final values under the name:")
df_all.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\edit\\%s.xlsx"%(q7d))
df_all_col_len = len(list(df_all.columns))
if df_all_col_len >=7:
    df_all_all = df_all.drop(df_all.columns[[1,2,4,5]], axis=1)
else:
    df_all_all = df_all.drop(df_all.columns[[1,2]],axis=1)



print(df_all_all)
ld = list(df_all_all.columns) 
ldsn = len(df_all_all[ld[0]])
sn = []
SN = list_append(sn, ldsn)
df_all_all.insert(0,"SN:", SN)
print(df_all_all)


q11A = input("Do you want to concat these files to the existing file?")
q11A = q11A.lower()
if q11A == 'y':
    q11B = input("Enter the name of the file with which you want to concat: ")
    q11C = pd.read_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(q11B))
    concated_file = pd.concat([q11C, df_all_all])
    print("Do you want to update the file->%s ?"%(q11B))
    q11D = input("Press 'y' or 'n'")
    q11D = q11D.lower()
    if q11D == 'y':
        list_col_name = list(concated_file.columns)
        row_len = len(concated_file[list_col_name[1]])
        sn = []
        SN= list_append(sn,row_len)
        concated_file = concated_file.drop(list_col_name[0], axis=1)
        concated_file.insert(0, "SN:" , SN)
        concated_file.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(q11B), index =False)
        concated_file.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\edit\\%s_modify2.xlsx"%(q11B))
    elif q11D == 'n':
        list_col_name = list(concated_file.columns)
        row_len = len(concated_file[list_col_name[1]])
        sn = []
        SN= list_append(sn,row_len)
        concated_file = concated_file.drop(list_col_name[0], axis=1)
        concated_file.insert(0, "SN:" , SN)
        q11E = input("Save the file under the name: ")
        concated_file.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(q11E), index =False)
        concated_file.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\edit\\%s_modify.xlsx"%(q11E))
if q11A == 'n':
    
    save_name = input("Save a file under the name: ")
    df_all_all.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\edit\\%s_modify.xlsx"%(save_name))
    df_all_all.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(save_name), index = False)


    
        
```
