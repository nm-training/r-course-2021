```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
q1 = int(input("Enter the total number of standard: "))
std_abs = []
std_conc = []
for i in range(q1):
    q2 = float(input("Enter the standard concentration: "))
    q3 = float(input("Enter the absorbance: "))
    std_abs.append(q3)
    std_conc.append(q2)
std_absorbance = np.array(std_abs).reshape((-1,1))
std_concentration = np.array(std_conc)
lr = LinearRegression()
lr.fit(std_absorbance, std_concentration)
r_sq = lr.score(std_absorbance,std_concentration)
print("Coefficient of determination for this graph is: ", r_sq)
print("y-intercept: ", lr.intercept_)
print("slope: ", lr.coef_)
print("equation of the line: ", "y=", lr.coef_[0],"*x+",lr.intercept_)

def ammonia_calculation(sample_abs):
    
    
    predicted_value = lr.coef_[0]*sample_abs + lr.intercept_
    
    final_value = (predicted_value/50)*(17/14)
    return final_value
def s_num(s_val, loop_val):
    s_val=[]
    for abc in range(loop_val):
        s_val.append(abc+1)
    return s_val

ans = input("Do you want to import the sample names? Press 'Y' or 'N' : ")
ans = ans.lower()

if ans == 'y':
    
    q6 = input("Enter file name from which you want to import: ")
    
    i_list = pd.read_excel('C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx'%(q6))
    
    i_val = list(i_list["Sample:"])
    len_i_val = len(i_val)
    print(" ")
    print (" ")
    SN = []
    SN = s_num(SN , len_i_val)
    dict1={}
    dict2={}
    dict3={}
    dict4={}
    for items in range(len_i_val):
        q= i_val[items]
        print(" ")
        print(" ")
        print("For sample:", "\n",  "%s"%(q))
        r= float(input("Enter sample absorbance:"))
        amm = ammonia_calculation(r)

        dict1[q] = amm
        dict2[q] = r
        dict3[q] = q
        dict4[q] = SN[items]

    print(dict1)
    ser1 = pd.Series(dict1)
    ser2 = pd.Series(dict2)
    ser3 = pd.Series(dict3)
    ser4 = pd.Series(dict4)

    d_f = pd.DataFrame({'SN:':ser4,'Sample:':ser3,'Absorbance:':ser2,'Ammonia:':ser1})
    print(d_f)
    
elif ans == 'n':
    
    q4 = int(input("Enter the number of samples: "))
    SN = []
    SN = s_num(SN , q4)
    dict1={}
    dict2={}
    dict3={}
    dict4={}
    for i in range(int(q4)):
        q= (input("Enter the name of sample: ")).upper()
        r= float(input("Enter sample absorbance:"))
        amm = ammonia_calculation(r)

        dict1[q] = amm
        dict2[q] = r
        dict3[q] = q
        dict4[q] = SN[i]

    print(dict1)
    ser1 = pd.Series(dict1)
    ser2 = pd.Series(dict2)
    ser3 = pd.Series(dict3)
    ser4 = pd.Series(dict4)

    d_f = pd.DataFrame({'SN:':ser4,'Sample:':ser3,'Absorbance:':ser2,'Ammonia:':ser1})
    print(d_f)
d_f2 = d_f.drop(['Absorbance:'], axis =1)
q11A = input("Do you want to concat these files to the existing file?")
q11A = q11A.lower()
if q11A == 'y':
    q11B = input("Enter the name of the file with which you want to concat: ")
    q11C = pd.read_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(q11B))
    concated_file = pd.concat([q11C, d_f2])
    print("Do you want to update the file->%s ?"%(q11B))
    q11D = input("Press 'y' or 'n'")
    q11D = q11D.lower()
    if q11D == 'y':
        list_col_name = list(concated_file.columns)
        row_len = len(concated_file[list_col_name[1]])
        sn = []
        SN= s_num(sn,row_len)
        concated_file = concated_file.drop(list_col_name[0], axis=1)
        concated_file.insert(0, "SN:" , SN)
        concated_file.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(q11B), index =False)
        concated_file.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\edit\\%s_modify2.xlsx"%(q11B))
    elif q11D == 'n':
        list_col_name = list(concated_file.columns)
        row_len = len(concated_file[list_col_name[1]])
        sn = []
        SN= s_num(sn,row_len)
        concated_file = concated_file.drop(list_col_name[0], axis=1)
        concated_file.insert(0, "SN:" , SN)
        q11E = input("Save the file under the name: ")
        concated_file.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(q11E), index =False)
        concated_file.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\edit\\%s_modify.xlsx"%(q11E))
if q11A == 'n':
    
    save_name = input("Save a file under the name: ")
    d_f.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\edit\\%s_modify.xlsx"%(save_name))
    d_f2.to_excel("C:\\Users\\Sanjay\\.anaconda\\office\\%s.xlsx"%(save_name), index = False)


    
```
